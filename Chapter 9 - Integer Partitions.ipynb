## Chapter 9: Integer Partitions

This is the Sage notebook for [**Chapter 9 of enumeration.ca**](https://enumeration.ca/extensions/partitions/) by Stephen Melczer. We examine how to use Sage to study integer partitions and their statistics.

```python
# Recall the class of partitions of size n
Partitions(5)
```

```python
# We can take random elements
Partitions(5).random_element()
```

```python
# We can draw diagrams
ascii_art(Partitions(100).random_element())
```

```python
# Enumerate partitions
[Partitions(n).cardinality() for n in range(10)]
```

```python
# You can put restrictions when generating the partitions
# For instance, you can restrict what numbers are allowed as parts
[len(Partitions(n, parts_in=[1,2,3])) for n in range(10)]
```

```python
# This matches the expected GF
A.<z> = QQ[['z']]
1/((1-z)*(1-z^2)*(1-z^3) + O(z^10))
```

```python
# Verify our asymptotic formula for these partitions
n = 200
(Partitions(n, parts_in=[1,2,3]).cardinality() / (n^2/12)).n()
```

```python
Pbox = Partitions(20, max_part=10, max_length=5)
Pbox.cardinality()
```

```python
ascii_art(Pbox.random_element())
```

```python
# We can define partitions with distinct parts using the "slope" between rows
[Partitions(n, max_slope=-1).cardinality() for n in range(10)]
```

```python
# Check this matches partitions with odd parts
n = 10
odds = [i for i in range(1, n + 1, 2)]
[Partitions(n, parts_in=odds).cardinality() for n in range(10)]
```

```python
# Sage can compute things like partition conjugates
L = Partitions(30).random_element()
show(ascii_art(L), ascii_art(L.conjugate()))
```
