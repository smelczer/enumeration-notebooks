## Chapter 12: Rational Asymptotics

This is the Sage notebook for [**Chapter 12 of enumeration.ca**](https://enumeration.ca/asymptotics/rational-asymptotics/) by Stephen Melczer. We examine how to manipulate rational generating functions in Sage and compute asymptotics of C-finite sequences.

```python
# Sage implements the ring of C-Finite sequences. One can imagine a recurrence and initial terms
# (or the corresponding rational GF) as providing an encoding of the sequence
C.<x> = CFiniteSequences(QQ)
print(C)
```

```python
# C-finite sequences can be defined by their recurrence and initial conditions
# (run C.from_recurrence? to see the format for the input)
fib = C.from_recurrence([1,1],[1,1]) # Recurrence a_{n+2} = a_{n+1} + a_n with a_0 = a_1 = 1
print(fib)
print(fib.parent())
```

```python
# C-finite sequences can also be defined by their rational generating function
fib2 = C('1/(1-x-x^2)')
print(fib2)
print(fib==fib2)
```

```python
# Power series truncation can be computed automatically
fib.series(10)
```

```python
# Coefficients can be accessed in a natural way
print([fib[k] for k in range(10)])
print(fib[1000])
```

```python
# We can get the closed from expression from the C-Finite Coefficient Theorem
show(fib.closed_form())
```

```python
# In general, the result cannot contain numbers explicitly defined with radicals
expr = C(1/(1-x-x^2-x^5)).closed_form()
show(expr)
```

```python
# Each term can be accessed individually
expr.operands()[0]
```

```python
# And the constants involved are actually exact algebraic numbers
power = expr.operands()[0].operands()[0].operands()[0]
power
```

```python
power.minpoly()
```

```python
power.n(200)
```

```python
# Given initial terms, Sage can also guess a C-finite sequence with those terms
# (see C.guess? for details on how this is done, checks to eliminate false positives, etc.)
C.guess([1,1,2,3,5,8])
```

```python
# Instead we can easily compute asymptotics. Note that we can compute the roots of H quickly
print(H.roots(QQbar))
```

```python
# We can verify simple roots without computing them using the following
H.diff().gcd(H)
```

```python
# Find minimal modulus of a root
rho = min([abs(r) for r in H.roots(QQbar, multiplicities=False)])
rho
```

```python
# Find all roots of minimal modulus -- there is only one (which is positive and real)
[r for r in H.roots(QQbar,multiplicities=False) if abs(r) == rho]
```

```python
# Coefficient in asymptotics
var('n')
L = (-G/z/H.derivative(z)).subs(z=rho)
L * (1/rho)^n
```

```python
# The coefficient appearing is algebraic of degree 71
L.minpoly().degree()
```
