## Chapter 10: Labelled Constructions 

This is the Sage notebook for [**Chapter 10 of enumeration.ca**](http://enumeration.ca/extensions/labelled-constructions/) by Stephen Melczer. We examine how to use compute generating functions for labelled objects using Sage's support for Species.

```python
# After translating to EGF equations we can use Power Series as in the unlabelled case
# Sage also handles labelled enumeration using "Species Theory"
from sage.combinat.species.library import SingletonSpecies, EmptySetSpecies, CombinatorialSpecies, SetSpecies, CycleSpecies 
Z = SingletonSpecies() # Atomic Class
E = EmptySetSpecies()  # Neutral Class
```

```python
# Define the SET and CYC operators
SET = SetSpecies()
CYC = CycleSpecies()
```

```python
# Permutations are sets of disjoint cycles
Perm = SET(CYC)
Perm
```

```python
# EGF of permutations is 1/(1-z)
Perm.generating_series()
```

```python
# Sage can also compute the GF of non-isomorphic objects (up to permuting labels)
Perm.isotype_generating_series()
```

```python
# Class of derangements
CYC_no_fixed = CYC.restricted(min=2)
Derangements = SET(CYC_no_fixed)
Derangements
```

```python
# Class of set partitions
SET_non_empty = SetSpecies().restricted(min=1)
SetPartitions = SetSpecies()(SET_non_empty)
SetPartitions
```

```python
# Enumerate set partitions
[factorial(n) * c for (c,n) in zip(SetPartitions.generating_series()[:10],range(10))]
```

```python
# Define rooted labelled trees
T = CombinatorialSpecies()
T.define(Z * SET(T))
[factorial(n) * c for (c,n) in zip(T.generating_series()[:10],range(10))]
```

```python
# Define functional mappings
K = CYC(T)
M = SET(K)
M
```

```python
# Enumerate functional mappings
[factorial(n) * c for (c,n) in zip(M.generating_series()[:10],range(10))]
```

```python
# As expected this is just n^n
[1] + [n^n for n in range(1,10)]
```

```python
# Enumerate functions with no fixed points
M_no_fixed = SET(CYC_no_fixed(T))
[factorial(n) * c for (c,n) in zip(M_no_fixed.generating_series()[:10],range(10))]
```

```python
# This is just (n-1)^n
[(n-1)^n for n in range(10)]
```

```python
# EGF for idempotent mappings
SET(Z * SET(Z)).generating_series()
```
