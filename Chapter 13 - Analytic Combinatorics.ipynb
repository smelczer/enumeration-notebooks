## Chapter 13: Analytic Combinatorics

This is the Sage notebook for [**Chapter 13 of enumeration.ca**](https://enumeration.ca/asymptotics/rational-asymptotics/) by Stephen Melczer. We examine how to use compute asymptotic expansions from generating functions in Sage.

```python
# The AsymptoticRing structure can compute asymptotic expansions
A = AsymptoticRing(
    growth_group='QQ^n * n^QQ * log(n)^QQ',
    coefficient_ring=SR,
    default_prec=7,  # default precision
)
n = A.gen()
A
```

```python
# It can do basic algebra / expansions
1/(1+n)
```

```python
exp(1-1/n)
```

```python
log(1+n)
```

```python
# It can compute the asymptotic contribution of a singularity 
# First we do the Catalan numbers
def f1(t): return (1-sqrt(1-4*t))/(2*t)

show(A.coefficients_of_generating_function(function=f1, singularities=(1/4), precision=4))
```

```python
# Next we do the Harmonic numbers
A.coefficients_of_generating_function(
    function=lambda z: -log(1 - z)/(1 - z),
    singularities=(1,),
    precision=10
)
```

```python
# Certain classes are built-in
asymptotic_expansions.HarmonicNumber('n', precision=5)
```

```python
# Asymptotics of surjections (the package does not detect there is an "exponentially small error")
show(A.coefficients_of_generating_function(
    function=lambda z: 1/(exp(z)-2),
    singularities=(log(2),),
    precision=10
))
```
