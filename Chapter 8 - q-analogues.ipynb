## Chapter 8: q-Analogues

This is the Sage notebook for [**Chapter 8 of enumeration.ca**](https://enumeration.ca/extensions/q-analogues/) by Stephen Melczer. We examine how to use Sage to study q-Analogues of combinatorial objects.

```python
# Import functions
from sage.combinat.q_analogues import q_int, q_factorial, q_binomial, q_catalan_number
var('q')
```

```python
# We start with q-numbers
q_int(3)
```

```python
q_int(10)
```

```python
# Next we define q-factorials
mul([q_int(k) for k in range(1,6)])
```

```python
# Built-in command for q-factorials
q_factorial(5)
```

```python
# Next we define q-binomial coefficients
q_factorial(5)/q_factorial(3)/q_factorial(2)
```

```python
# Built-in command for q-binomials
q_binomial(5,3)
```

```python
# Rows in "q-Pascal Triangle"
for n in range(5):
    show([q_binomial(n,k) for k in range(n+1)])
```

```python
# Generating function for permutations of size n by inversions
n = 6
add([q^len(pi.inversions()) for pi in Permutations(n)])
```

```python
# This matches the q-factorial
q_factorial(n)
```

```python
# Verify q-binomial theorem
# Enumerate subsets of [n] of size k by sum
n = 5
k = 3
add([q^add(S) for S in Subsets(n) if len(S) == k])
```

```python
# This matches the scaled q-binomial coefficient
(q^(k*(k+1)/2) * q_binomial(n,k)).expand()
```

```python
# Verify q-binomial formula for lattice paths
# Note: lattice paths from (0,0) to (a,b) are in bijection with partitions in an a x b box
# (Move from the bottom left corner of the box to the top right, staying below the Young Diagram)

a,b = 3,2
add([q^(a*b-sum(k)) for k in PartitionsInBox(a,b)])
```

```python
q_binomial(a+b,a)
```

```python
# We can verify q-analogue formulas experimentally
n = 5
k = 3
print(q_binomial(n, k))
print((q_binomial(n-1, k) + q^(n-k)*q_binomial(n-1, k-1)).expand())
```

```python
# We can also explore new q-analogues of more complicated sequences
# First, we define the "MacMahon" q-Catalan number
n = 5
1/q_int(n+1) * q_binomial(2*n,n)
```

```python
# Built-in command for MacMahon q-Catalan number in Sage
q_catalan_number(n)
```

```python
# Catalan numbers enumerate Dyck Paths
# What statistic can be use for the q-analogue?

# The "major index" of a Dyck path is the sum of the indices of the D steps in its "valleys" DU
dw = DyckWords(5)[10]
dw
```

```python
# Note: major_index starts counting at index 1 while valleys starts at index 0
print(dw.valleys())
print(dw.major_index())
```

```python
add([q^w.major_index() for w in DyckWords(n)])
```

```python
# There are other natural q-analogues of the Catalan numbers that track different parameters!
# For instance, the Catalan numbers satisfy C(n) = C(0)C(n-1) + C(1)C(n-2) + ... + C(n-1)C(0)
# We can generalize this to the "Carlitz" q-binomial coefficient

@cached_function
def carlitz_q_catalan(n):
    if n == 0: return 1
    else:      return sum([q^k * carlitz_q_catalan(k) * carlitz_q_catalan(n - k - 1) 
               for k in range(n)]).expand()
```

```python
n = 6
carlitz_q_catalan(n)
```

```python
# This tracks Dyck paths by area!
add([q^w.area() for w in DyckWords(n)])
```

```python
# This corresponds to "sum of left-subtree size" in binary trees
def left_sizes(t):
    if t.is_empty(): return 0
    return t[0].node_number() + left_sizes(t[0]) + left_sizes(t[1])
```

```python
n=5
add([q^left_sizes(t) for t in BinaryTrees(n)])
```
