```python
# We can also build systems of equations 
# Here we study rooted planar trees whose internal vertices at even height have an even number of children
# and whose internal vertices at odd height have an odd number of children.
E  = S.undefined()
O  = S.undefined()
E.define(x/(1-O^2))
O.define(x + x*E/(1-E^2))
```

```python
# Unfortunately, LazyPowerSeries needs to explicitly see the denominator has the form 1-xF for a series F
E
```

```python
# To help LazyPowerSeries we specify A = E/x and B = O/x
A = S.undefined()
B = S.undefined()
A.define(1 / (1 - (x*B)^2))
B.define(1 + x*A / (1 - (x*A)^2))

# Then recover the actual generating functions E and O
E = x * A
O = x * B
```

```python
# First 10 terms of the counting sequence
E[:10]
```

```python
# Number of such trees with 200 vertices
E[200]
```

```python
# Counting sequence where internal vertices at even height have an odd number of children (and vice-versa)
O[:10]
```
